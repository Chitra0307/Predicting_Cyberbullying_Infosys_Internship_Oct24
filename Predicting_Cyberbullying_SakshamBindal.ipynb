## Removing unneccesory spaces and replacing @ with a and h


```python
import pandas as pd
import re

# Define a function to process text: removing URLs, unnecessary spaces, and replacing special characters
def process_text(text):
    # Eliminate URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Substitute '@' with 'a' and '#' with 'h'
    text = text.replace('@', 'a').replace('#', 'h')

    # Remove excess whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Assuming your DataFrame is named 'data'
# Example: data = pd.read_csv('your_dataset.csv')

# Apply the processing function to clean the 'tweet_text' column
data['processed_tweet'] = data['tweet_text'].apply(process_text)

# Display the original and cleaned data for comparison
print(data[['tweet_text', 'processed_tweet']].head())
data.head(20)
```
